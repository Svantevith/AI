```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.applications import imagenet_utils
from sklearn.metrics import confusion_matrix
import itertools
import os
import shutil
import random
import matplotlib.pyplot as plt
from IPython.display import display, Image
from image_set import prepare_image
```

```python
mobile_model = tf.keras.applications.mobilenet.MobileNet()
```

```python
src_path = 'data/MobileNet Samples/'
file_name_1 = 'lizard.png'
```

```python
img_1 = Image(filename=src_path+file_name_1)
display(img_1)
```

```python
preprocessed_image_1 = prepare_image(file_name_1)
```

```python
predictions_1 = mobile_model.predict(preprocessed_image_1)
```

```python
results_1 = imagenet_utils.decode_predictions(predictions_1)
```

```python
results_1
```

```python
assert results_1[0][0][1] == 'American_chameleon'
```
